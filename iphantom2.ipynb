Use original phantom

```python
radius = 37.0
well_seps = (8.0, 6.0, 5.0, 4.0, 3.0, 2.0)
test_phantom = DerenzoPhantom(radius, well_seps)
test_phantom.show()
```

Generate Mask for reference phantom

```python
myMask=test_phantom.mask.reshape(1024,1024)
ic(myMask.shape)
ic(myMask.sum())
plt.imshow(myMask)
```
